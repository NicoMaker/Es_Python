# MINIMIZZAZIONE DI UN BUDGET CON LA SOMMA DI DUE NUMERI UGUALI A 3M

```python
# dicharazione costanti

rischioA = 0.3
rischioB = 0.2
budget = 3_000_000
x = budget

# calcolo del rischio totale

for progettoB in range(budget + 1):

    rischio_tot = (budget - progettoB) * rischioA + progettoB * rischioB

    if rischio_tot < x:

        x = rischio_tot
        progettoAX = budget - progettoB
        progettoBX = progettoB

else:

    print(
        f"il valore minimo aggiunto  : {x}  con rispettivi valori dei progetti")
    print(f"ProgettoA {progettoAX} , ProgettoB : {progettoBX}")
```

# MINIMIZZAZIONE DI UN BUDGET CON SOMMA DI NUE NUMERI MINORI UGUALI A 3M

```python
# dicharazione costanti

rischioA = 0.3
rischioB = 0.2
budget = 3_000_000
x = budget

# calcolo del rischio totale

for progettoB in range(budget + 1):

    for progettoA in range(progettoB):

        rischio_tot = (budget - progettoB) * rischioA + progettoB * rischioB

        if rischio_tot < x:

            x = rischio_tot
            progettoAX = budget - progettoB
            progettoBX = progettoB

else:

    print(
        f"il valore minimo aggiunto  : {x}  con rispettivi valori dei progetti")
    print(f"ProgettoA {progettoAX} , ProgettoB : {progettoBX}")
```
