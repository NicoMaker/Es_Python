### 2. Definisci un inizializzatore per la classe con tre parametri: "nome", "conto", "importo"
### 3. La classe deve avere tre attributi di istanza: "nome", "conto", "saldo"
### 4. Definisci un metodo di istanza denominato "preleva" con parametro "importo" che diminuisce il valore del "saldo" del conto del valore "importo"
### 5. Definisci un metodo di istanza denominato "depostia" con parametro "importo" che aumenta il valore del "saldo" del conto del valore "importo"
### 6. Definisci un metodo di istanza "descrizione" che visualizza "nome", "conto", "saldo" dell'istanza.

```python
# 1. Definisci una classe ContoCorrente

class ContoCorrente:
    
    # 2. Definisci un inizializzatore per la classe con tre parametri: nome, conto, importo
    # 3. La classe deve avere tre attributi di istanza: nome, conto, saldo
    
    def __init__(self, nome, conto, importo):
        self.nome = nome
        self.conto = conto
        self.saldo = importo
        
        print(f"E' stato creato il conto {self.conto} a nome di {self.nome} per un importo iniziale di {self.saldo}€")
        print("")
        
    # 4. Definisci un metodo di istanza "preleva" con parametro "importo" che non ha valore di ritorno
    # e diminuisce il valore del saldo di quell'importo
    
    def preleva(self, importo):
        self.saldo -= importo
        
        print(f"E' stato prelevato l'importo di {importo}€ dal conto {self.conto}")
        print(f"il saldo attuale è di: {self.saldo}€")
        print("")
    
    # 5. Definisci un metodo di istanza "deposita" con parametro "importo" che non ha valore di ritorno
    # ed aumenta il valore del saldo di quell'importo
    
    def deposita(self, importo):
        self.saldo += importo
        
        print(f"E' stato depositato l'importo di {importo}€ dal conto {self.conto}")
        print(f"il saldo attuale è di: {self.saldo}€")
        print("")
    
    # 6. Definisci un metodo di istanza chiamato "descrizione" che visualizza nome, conto e saldo
    
    def descrizione(self):
        
        print(f"Il conto {self.conto} di {self.nome} ha un saldo di {self.saldo}€")
        print("")

# Inizializzo tre istanze

c1 = ContoCorrente("Cristian", "0001A", 15000)
c2 = ContoCorrente("Tristano", "0002A", 1000)
c3 = ContoCorrente("Ubaldo", "0003C", 7000)

# Test del metodo preleva
c1.preleva(2000)

# Test del metodo deposita
c2.deposita(1350)

# Test del metodo descrizione
c1.descrizione()
c2.descrizione()
c3.descrizione()
```

## SECONDA PARTE

### 1. Creare la classe Conto
### 2. Dentro Conto le istanze devono avere due attributi: nome, conto che devono essere richieste in fase di inizializzazione
### 3. Modifica la Classe ContiCorrenti in modo che sia sottoclasse di Conto
### 4. Modificare l'inizializzatore di ContiCorrenti facendo in modo che nome e conto siano ereditati dalla superclasse Conto
### 5. Nascondi l'attributo saldo
### 6. Definisci mediante decoratori get e set per l'attributo saldo con nome saldo
### 7. Get deve poter far vedere il valore del saldo
### 8. Set deve settare il nuovo valore inserito come segue: usare il metodo preleva prelevando il saldo attuale e usare il metodo deposita con il valore da settare.

-----------------------------------------------------------------------------------------

```python
class Conto:
    def __init__(self, nome, conto):
        self.nome = nome
        self.conto = conto
        
class ContoCorrente(Conto):
    
    # 1. Nascondi l'attributo "saldo"
    
    def __init__(self, nome, conto, importo):
        super().__init__(nome, conto)
        self.__saldo = importo
        
        print(f"E' stato creato il conto {self.conto} a nome di {self.nome} per un importo iniziale di {self.__saldo}€")
        print("")
        
    # 2. Definisci una property (mediante decorator) GETTER di nome saldo che ritorna l'attributo "saldo"
    
    def preleva(self, importo):
        self.__saldo -= importo
        
        print(f"E' stato prelevato l'importo di {importo}€ dal conto {self.conto}")
        print(f"il saldo attuale è di: {self.__saldo}€")
        print("")
    
    def deposita(self, importo):
        self.__saldo += importo
        
        print(f"E' stato depositato l'importo di {importo}€ dal conto {self.conto}")
        print(f"il saldo attuale è di: {self.__saldo}€")
        print("")
    
    @property
    def saldo(self):
        print("-- Invoca il GETTER di saldo --")
        return self.__saldo
    
    # 3. Definisci una property (mediante decorator) SETTER di nome saldo che:
    #       a. Invoca il metodo "preleva" con importo pari al saldo attuale
    #       b. Invoca il metodo "deposita" con importo definito come parametro
    
    @saldo.setter
    def saldo(self, importo):
        print("-- Invoca il SETTER di saldo --")
        self.preleva(self.__saldo)
        self.deposita(importo)
        
    def descrizione(self):
        
        print(f"Il conto {self.conto} di {self.nome} ha un saldo di {self.__saldo}€")
        print("")

# Inizializzo tre istanze

c1 = ContoCorrente("Cristian", "0001A", 15000)
c2 = ContoCorrente("Tristano", "0002A", 1000)
c3 = ContoCorrente("Ubaldo", "0003C", 7000)

# Test del metodo preleva
c1.preleva(2000)

# Test del metodo deposita
c2.deposita(1350)

# Test GETTER

c1.saldo

# Test SETTER

c1.saldo = 80000

# Test del metodo descrizione
c1.descrizione()
c2.descrizione()
c3.descrizione()
```

```python
# 1. Creare la classe GestioneContiCorrenti

class GestioneContiCorrenti:
    
# 2. Creare un metodo di classe all'interno di GCC che funzione come 
# di seguito specificato:
#           a. Il nome del metodo è "bonifico"
#           b. E' un metodo statico
#           c. I parametri sono "sorgente", "destinazione", "importo"
#           d. Invoca il metodo "preleva" sul conto "sorgente" con valore 
#              "importo"
#           e. Invoca il metodo "deposita" sul conto "destinatario" con
#              valore "importo"

    @staticmethod
    def bonifico(sorgente, destinazione, importo):
        sorgente.preleva(importo)
        destinazione.deposita(importo)        
        print(f"Bonifico dal conto {sorgente.conto} al conto {destinazione.conto} di importo {importo} eseguito!")
        print(f"I nuovi importi sono: ")
        print(f"{sorgente.conto}: Saldo: {sorgente.saldo}")
        print(f"{destinazione.conto}: Saldo: {destinazione.saldo}")
        print("")
        
# Main

# Inizializzo 3 istanze differenti

c1 = ContoCorrente("Cristian", "0001A", 15000)
c2 = ContoCorrente("Tristano", "0002A", 5000)
c3 = ContoCorrente("Ubaldo", "0017C", 85000)

# 3. Esegui un bonifico tra due conti

GestioneContiCorrenti.bonifico(c1, c2, 3_000)
```
